```python
import matplotlib.pyplot as plt 
```

```python
x = ["a", "b", "c", "d", "e"]
```

```python
y = [5, 4, 3, 12, 1]
```

```python
plt.bar(x , y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("bar plot")
```

```python
plt.scatter(x , y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("bar plot")
```

```python
plt.plot(x , y , '--b')
plt.xlabel("This is my x data")
plt.ylabel("This is my y data")
plt.title("This is my title")
plt.grid()
plt.show()
```
